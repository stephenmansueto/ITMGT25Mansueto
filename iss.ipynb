```python
import requests
import time

data = []
for i in range (150):
    res = requests.get('http://api.open-notify.org/iss-now.json')
    json_data = res.json()
    data.append(json_data)
    time.sleep(10)
    
```

```python
data
```

```python
headers = ['timestamp', 'latitude', 'longitude']

body = []
for record in data:
    row = []
    row.append(record['timestamp'])
    row.append(record['iss_position']['latitude'])
    row.append(record['iss_position']['longitude'])
    body.append(row)
```

```python
body
```

```python
len(body)
```

```python
import matplotlib.pyplot as plt

datalist = list(zip(*body))

x = datalist[1]
y = datalist[2]

plt.xlabel('latitude')
plt.ylabel('longitude')
plt.title('ISS dataplots')

plt.plot(x,y, color = '#4169e1')
plt.show()
```
